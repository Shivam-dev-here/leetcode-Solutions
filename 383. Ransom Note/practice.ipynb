# With Counter

```python
from collections import Counter

ransomNote = "aba"
magazine = "aaaab"

r = Counter(ransomNote)
m = Counter(magazine)

print(r,m)

for ch in r:
    if r[ch] > m[ch]:
        print("False")
        break
else:
    print("True")
```

```python
# Counter subtraction removes negative values & empty counter is True
from collections import Counter

ransomNote = "aba"
magazine = "ab"


print(not (Counter(ransomNote) - Counter(magazine)))
```
